# RESULTADOS ENTRENAMIENTOS

Los entrenamientos se han realizado con ventanas deslizantes sobre todo el conjunto de datos del abastecimiento de Castellón. La ventana de train ocupa 365 días (1 año), mientras que con la ventana de test se han hecho dos pruebas diferentes, ocupando 30 días, es decir un més y 7 días (una semana).

Anteriormente se probaron diferentes modelos (con solo dos clusters de prueba) para posteriormente elegir aquellos con mejores resultados. Los elegidos se han entrenado sobre todos los clusters.

#### Métricas utilizadas:
- MAE (Error Absoluto Medio): mide el error promedio en las predicciones, sin tener en cuenta la dirección del error.
- RMSE (Raíz del Error Cuadrático Medio): pondera los errores grandes más que los pequeños, siendo más sensible a valores atípicos.
- MAPE (Error Absoluto Medio Porcentual): mide el error en porcentaje, lo que lo hace independiente de la escala de los datos.

**Observaciones y Análisis**

- LSTM domina en precisión:

LSTM es el mejor modelo en todos los clústeres, sin embargo hay que tener en cuenta que tratar los datos con él es más complicado y que su entrenamiento ha sido un poco diferente,
la ventana era de 365 con un input de 30 días de los que se iba prediciendo día a día, lo cual altera en su favor un poco los resultados respecto al resto.

Holt-Winters ofrece un desempeño sólido:

- Holt-Winters es competitivo en algunos clústeres, especialmente en los clústeres 4 y 9, donde se acerca al desempeño de LSTM.
Dado que es un modelo más simple, es sorprendente que mantenga un nivel de precisión relativamente cercano en ciertos casos.
NeuralProphet muestra limitaciones:

- NeuralProphet las mejoras realizadas sobre este modelo, tanto en parámetros como en los datos hacen que los resultados bajen y se asemejen más a los de LSTM. Su tratamiento en general es más sencillo, y realmente ya combina modelos de predicción normales con redes neuronales, lo que lo convierte en un híbrido.

**Recomendaciones**
1. Usar LSTM para aplicaciones críticas:

Para casos donde la precisión es clave y los recursos computacionales no son una limitación, LSTM sería una buena opción.

2. Considerar Holt-Winters para simplicidad:

Si se busca un balance entre simplicidad y precisión, Holt-Winters es adecuado para la mayoría de los clústeres, especialmente los clústeres con menor variabilidad (como 0, 4, y 9).

3. Mejorar NeuralProphet:

Aunque tiene un desempeño un tanto inferior, NeuralProphet podría mejorarse ajustando hiperparámetros o añadiendo variables adicionales en el modelo.
